```python
import matplotlib.pyplot as plt

months =  ["Jan","Feb","mar","apr","may","Jun"]

sales = [10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5)) # Adjust figsize to accommodate three plots

# Line Chart
plt.subplot(1,3,1) # Changed to 1 row, 3 columns,position 1

plt.plot(months,sales,marker="o")

plt.title("Sales Trend")

plt.xlabel("Months")

plt.xlabel("Sales")

#Bar chart

plt.subplot(1,3,2) # Changed to 1 row, 3 columns, position 2

plt.bar(months,

sales,

color="orange")

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

#Pie chart

plt.subplot(1,3,1) # Changed to 2 row, 3 columns, position 3

expenses = [5000, 3000, 2000]

categories = ["Food", "Travel", "Shopping"]

plt.pie(expenses,

labels=categories, autopct="%1.1f%%", colors = ["pink", "yellow", "green"])

plt.title("Expense Distribution")

plt.tight_layout()

plt.show()
```
