STUDENT SCORE ANALYSIS

```python
import numpy as np

scores = np.array([85, 90, 78, 92, 88])

print("Student Scores:", scores)
print("Total Score:", np.sum(scores))
print("Average Score:", np.mean(scores))
print("Highest Score:", np.max(scores))
print("Lowest Score:", np.min(scores))
```

```python
import numpy as np

sales = np.array([1200, 1500, 1800, 2000, 1700])

print("Sales Data:", sales)
print("Total Sales:", np.sum(sales))
print("Average Sales:", np.mean(sales))
print("Maximum Sales:", np.max(sales))
print("Minimum Sales:", np.min(sales))
```

PROTECT SALES PERFORMANCE ANALYSIS

```python
import pandas as pd

# Product sales data
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"],
    "Sales": [50000, 30000, 20000, 10000, 15000]
}

df = pd.DataFrame(data)

print("Product Sales Data")
print(df)

# Total sales
print("\nTotal Sales:", df["Sales"].sum())

# Average sales
print("Average Sales:", df["Sales"].mean())

# Highest selling product
print("Highest Selling Product:")
print(df.loc[df["Sales"].idxmax()])

# Lowest selling product
print("Lowest Selling Product:")
print(df.loc[df["Sales"].idxmin()])
```
