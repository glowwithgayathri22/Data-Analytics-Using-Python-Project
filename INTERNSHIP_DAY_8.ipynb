```python
import pandas as pd
data = {
    "Region":["North","South","North","East"],
    "Sales":[1000,2000,1500,3000]
}

df = pd.DataFrame(data)

region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)

```

```python

import pandas as pd
data = {
    "Product":["North","South","North","East"],
    "Sales":[1000,2000,1500,3000]
}
df = pd.DataFrame(data)

Product_Sales =df. groupby("Product")["Sales"].sum()
print(Product_Sales)
```

```python
import pandas as pd

data ={
"Region":["South", "North", "East", "West", "South", "North"],
"Product":["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Table"],
"Sales":[50000, 30000, 45000, 25000, 20000, 35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df["Sales"].sum())
```

```python
import matplotlib.pyplot as plt

month=["Jan", "Feb", "Mar", "Apr"," may"]

Sales = [20000, 30000,40000,50000,60000]

plt. plot (month, Sales)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales Amount")

plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["A", "B","C", "D"]

marks = [80,90,75,95]

plt.bar(students, marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```
