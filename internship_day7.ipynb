```python
import pandas as pd

data = {
 "Region": ["South", "North", "East", "West", "South", "North"],
 "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
 "Sales": [50000, 30000, 45000, 25000, 20000, 35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
south_sales = df[df["Region"] == "South"]

print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]

print(high_sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

=> CLEAN A MESSY SALES DATASET


```python
import pandas as pd
import numpy as np

# Messy dataset
data = {
    "Product": ["Laptop", "Mobile", "Tablet", None, "Laptop"],
    "Price": [50000, 20000, np.nan, 30000, 50000],
    "Quantity": [2, 3, 1, np.nan, 2]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing Product values
df["Product"] = df["Product"].fillna("Unknown")

# Fill missing Price with average price
df["Price"] = df["Price"].fillna(df["Price"].mean())

# Fill missing Quantity with 0
df["Quantity"] = df["Quantity"].fillna(0)

print("\nCleaned Dataset:")
print(df)
```
